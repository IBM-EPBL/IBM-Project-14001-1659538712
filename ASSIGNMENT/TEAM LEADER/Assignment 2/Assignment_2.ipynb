# ***6. Find the outliers and replace the outliers***

```python
sns.boxplot(data['Age'])
```

```python
sns.boxplot(data['Balance'])
```

```python
sns.boxplot(data['CreditScore'])
```

```python
sns.boxplot(data['Tenure'])
```

```python
sns.boxplot(data['EstimatedSalary'])
```

***The columns Age and CreditScore are having the outliers.***

***Detecting outliers Age column and removing outliers.***

```python
Q1 = np.percentile(data['Age'], 25,
                   interpolation = 'midpoint')
 
Q3 = np.percentile(data['Age'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1
 
print("Old Shape: ", data.shape)
 
# Upper bound
upper = np.where(data['Age'] >= (Q3+1.5*IQR))
# Lower bound
lower = np.where(data['Age'] <= (Q1-1.5*IQR))
```

```python
''' Removing the Outliers '''
data.drop(upper[0], inplace = True)
data.drop(lower[0], inplace = True)
 
print("New Shape: ", data.shape)
```

```python
sns.boxplot(data['Age'])
```

***Detecting outliers CreditScore column and removing outliers.***

```python
q1 = np.percentile(data['CreditScore'], 25,
                   interpolation = 'midpoint')
 
q3 = np.percentile(data['CreditScore'], 75,
                   interpolation = 'midpoint')
iqr = q3 - q1
 
print("Old Shape: ", data.shape)
 
# Upper bound
upper1 = np.where(data['CreditScore'] >= (q3+1.5*iqr))
# Lower bound
lower2 = np.where(data['CreditScore'] <= (q1-1.5*iqr))
```

```python
''' Removing the Outliers '''
data.drop(upper1[0], inplace = True)
data.drop(lower2[0], inplace = True)
 
print("New Shape: ", data.shape)
```
